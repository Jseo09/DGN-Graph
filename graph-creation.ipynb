# Step 2: Data Manipulation
Before this dataset is made into graphs, we first need to get our data optimized to work for this.

```python
# We first properly convert the timestamps to panda's native format.
df['Timestamp'] = pandas.to_datetime(df['Timestamp'], format='%d/%m/%Y %I:%M:%S %p')
# Then create a rounded to the minute field for use later.
df['timestamp_minute'] = df['Timestamp'].dt.floor('min')
df
```

Once we have our minute rounded field, we now want to focus on a small subsection of the dataset to test graph creation. Specifically, we will work with the first 5 unique minutes in the dataset.

```python
# Firstly, we sort the dataframe by timestamp earliest to latest...
df = df.sort_values(by=['Timestamp'])
df
```

```python
# Then we select the first 5 unique minutes in the dataset...
first_5_minutes = df['timestamp_minute'].drop_duplicates().head(5)
# and create a new dataset of just those entries.
filtered_df = df[df['timestamp_minute'].isin(first_5_minutes)]
filtered_df
```

```python
# And lastly, list the unique minute entries.
unique_minutes = filtered_df['timestamp_minute'].unique()
print(unique_minutes)
```

```python
# We also group together the entries by the minute to see how large each graph should be.
grouped = filtered_df.groupby('timestamp_minute')
counts = grouped.size()
print(counts)
```

```python
# And print the dataframe one last time.
filtered_df = filtered_df.sort_values(by='Timestamp', ascending=True)
print(filtered_df.to_string())
```

# Step 3: Graph Creation
Now that the dataframe has been properly filtered for testing purposes, I now will begin the creation of a graph.

```python
# For testing, we grab only the first group of entries.
first_group_key = list(grouped.groups.keys())[0]
first_group = grouped.get_group(first_group_key)
first_group
```

## Step 3.1: Graph Creation Details
For current testing, I am going to start by focusing on a basic implementation of the graphing logic set out by the paper. In the next section, I will focus on adding layer based graphs, but for testing purposes, the graph below will represent graphing an entire minute of traffic.

```python
# For testing, nodes will be named by their index.
index = 0
# Then, we create the graph.
G = nx.DiGraph()
# And maintain the list of nodes.
nodes_list = []
# Lastly, we establish a previous node variable to store for easier graph creation.
previous_backward_node = None
# Now, for each entry...
for idx, row in first_group.iterrows():
    # We record the source and destination IP, as well as the flow ID.
    src_ip = row['Src IP']
    dst_ip = row['Dst IP']
    flow_id = row['Flow ID']
    # Then, we grab the forward node features for the entry. (Note: the features collection are the ones in section 3 table 2 of the DGNN paper.)
    forward_features = {
        'total_fwd_pkts': row['Total Fwd Packet'],
        'fwd_pkts_per_s': row['Fwd Packets/s'],
        'fwd_iat_std': row['Fwd IAT Std'],
        'total_len_fwd_pkt': row['Total Length of Fwd Packet'],
        'fwd_pkt_len_std': row['Fwd Packet Length Std'],
        'fwd_seg_size_avg': row['Fwd Segment Size Avg'],
        'fwd_init_win_bytes': row['FWD Init Win Bytes'],
        'fwd_pkt_len_mean': row['Fwd Packet Length Mean'],
        'fwd_iat_max': row['Fwd IAT Max'],
        'avg_pkt_size': row['Average Packet Size'],
        'subflow_fwd_bytes': row['Subflow Fwd Bytes'],
        'ip': src_ip,
        'direction': "forward"
    }
    # We then build the node based on those features and add it to the graph.
    forward_node = f"{index}_fwd"
    G.add_node(forward_node, **forward_features)
    nodes_list.append(forward_node)
    # Rinse and repeat for the backwards nodes...
    backward_features = {
        'total_bwd_pkts': row['Total Bwd packets'],
        'bwd_pkts_per_s': row['Bwd Packets/s'],
        'bwd_iat_std': row['Bwd IAT Std'],
        'total_len_bwd_pkt': row['Total Length of Bwd Packet'],
        'bwd_pkt_len_std': row['Bwd Packet Length Std'],
        'bwd_seg_size_avg': row['Bwd Segment Size Avg'],
        'bwd_init_win_bytes': row['Bwd Init Win Bytes'],
        'bwd_pkt_len_mean': row['Bwd Packet Length Mean'],
        'bwd_iat_max': row['Bwd IAT Max'],
        'avg_pkt_size': row['Average Packet Size'],
        'subflow_bwd_bytes': row['Subflow Bwd Bytes'],
        'ip': dst_ip,
        'direction': "backward"
    }

    backward_node = f"{index}_bkd"
    G.add_node(backward_node, **backward_features)
    nodes_list.append(backward_node)
    # And connect the edges as appropriate.
    G.add_edge(forward_node, backward_node)
    if previous_backward_node is not None:
        G.add_edge(previous_backward_node, forward_node)
    previous_backward_node = backward_node

    index += 1
```

```python
# Here we make sure the node list was correctly made...
print("Nodes List:", nodes_list)
```

```python
# Here we make sure the nodes have the features properly added...
for node in G.nodes(data=True):
    print(node)
```

```python
# And for testing, we also will print the graph with matplotlib.
pos = nx.spring_layout(G) 
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
plt.title('Network Traffic Graph')
plt.show()
```
